```python
title = soup.find("div", attrs={"class":"_3wU53n"})
print(title.text)
```

```python
titles = soup.find_all("div", attrs={"class":"_3wU53n"})
for title in titles:
    print(title.text)
```

```python
cards = soup.find("div",attrs={"class":"_1UoZlX"})
title = card.find("div", attrs={"class":"_3wU53n"})
price = card.find("div",attrs={"class":"_2rQ-NK"})
rating = card.find("div",attrs={"class":"hGSR34"})
print("{} {} {}".format(title.text,price.text,rating.text))
```
